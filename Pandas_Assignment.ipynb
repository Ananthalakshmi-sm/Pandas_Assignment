60.Create a monthly revenue trend using resample() on OrderDate.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Convert OrderDate to datetime and ensure TotalAmount is numeric
data['OrderDate'] = pd.to_datetime(data['OrderDate'], errors='coerce')
data['TotalAmount'] = pd.to_numeric(data['TotalAmount'], errors='coerce')

# 2. Set OrderDate as index (required for resample)
monthly_revenue = data.set_index('OrderDate')['TotalAmount'].resample('M').sum()

# 3. Create DataFrame with clean monthly data
monthly_trend = pd.DataFrame({
    'YearMonth': monthly_revenue.index.strftime('%Y-%m'),
    'Revenue': monthly_revenue.values,
    'MonthName': monthly_revenue.index.strftime('%B'),
    'Year': monthly_revenue.index.year
})

# 4. Plot the trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_trend['YearMonth'], monthly_trend['Revenue'],
         marker='o', linestyle='-', color='steelblue')
plt.title('Monthly Revenue Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 5. Display the monthly data
print("Monthly Revenue Summary:")
print(monthly_trend.to_string(index=False))
```

61. Find the day of the week with the highest sales volume.

```python
import pandas as pd

# 1. Convert OrderDate to datetime and extract weekday
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['Weekday'] = data['OrderDate'].dt.day_name()  # e.g. 'Monday', 'Tuesday'

# 2. Calculate total sales by weekday
weekday_sales = (
    data.groupby('Weekday', as_index=False)['TotalAmount']
    .sum()
    .sort_values('TotalAmount', ascending=False)
)

# 3. Add custom weekday order for proper sorting
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_sales['Weekday'] = pd.Categorical(weekday_sales['Weekday'], categories=weekday_order, ordered=True)
weekday_sales = weekday_sales.sort_values('Weekday')

# 4. Find the top weekday
top_weekday = weekday_sales.nlargest(1, 'TotalAmount')
top_day = top_weekday['Weekday'].values[0]
top_sales = top_weekday['TotalAmount'].values[0]

# 5. Display results
print("Sales Volume by Weekday:")
print(weekday_sales.to_string(index=False))
print(f"\nThe highest sales day is {top_day} with ${top_sales:,.2f} in total sales")

# 6. Optional visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(weekday_sales['Weekday'], weekday_sales['TotalAmount'], color='brown')
plt.title('Total Sales by Day of Week', fontsize=14)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

62.Compare monthly returned orders count vs. total orders.

```python
import pandas as pd

# 1. Prepare the data
# Convert to datetime and ensure consistent status values
data['OrderDate'] = pd.to_datetime(data['OrderDate'], errors='coerce')
# Ensure 'Returned' column is boolean
data['Returned'] = data['Returned'].astype(bool)

# Drop rows where 'OrderDate' became NaT due to parsing errors
data = data.dropna(subset=['OrderDate']).copy()

# Reset index to make OrderID a column before grouping
data_reset = data.reset_index()

# Rename the column created by reset_index to 'OrderID' if it's named 'index'
# This handles cases where the original index didn't have a name.
if 'index' in data_reset.columns and 'OrderID' not in data_reset.columns:
    data_reset.rename(columns={'index': 'OrderID'}, inplace=True)

# 2. Create month-year column for grouping
data_reset['YearMonth'] = data_reset['OrderDate'].dt.to_period('M')

# 3. Calculate metrics
monthly_stats = (
    data_reset.groupby('YearMonth')
    .agg(
        TotalOrders=('OrderID', 'count'), # Now 'OrderID' should exist as a column
        ReturnedOrders=('Returned', lambda x: x.sum())
    )
    .reset_index()
)

# 4. Calculate return rate
monthly_stats['ReturnRate'] = (monthly_stats['ReturnedOrders'] / monthly_stats['TotalOrders']) * 100
monthly_stats['YearMonth'] = monthly_stats['YearMonth'].dt.strftime('%Y-%m')  # Convert to string

# 5. Find the month with highest return rate
max_return = monthly_stats.loc[monthly_stats['ReturnRate'].idxmax()]

# 6. Display results
print("Monthly Order Comparison:")
print(monthly_stats.to_string(index=False))
print(f"\nHighest return rate: {max_return['ReturnRate']:.2f}% in {max_return['YearMonth']}")

# 7. Visualization
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Total vs Returned Orders
ax1.bar(monthly_stats['YearMonth'], monthly_stats['TotalOrders'], label='Total Orders', alpha=0.7)
ax1.bar(monthly_stats['YearMonth'], monthly_stats['ReturnedOrders'], label='Returned Orders', color='red', alpha=0.7)
ax1.set_title('Monthly Orders: Total vs Returned')
ax1.set_ylabel('Order Count')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# Return Rate
ax2.plot(monthly_stats['YearMonth'], monthly_stats['ReturnRate'],
         marker='o', color='green', label='Return Rate (%)')
ax2.set_title('Monthly Return Rate')
ax2.set_ylabel('Return Rate (%)')
ax2.axhline(y=monthly_stats['ReturnRate'].mean(), color='r', linestyle='--',
            label=f'Avg: {monthly_stats["ReturnRate"].mean():.2f}%')
ax2.grid(axis='y', alpha=0.3)
ax2.legend()
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

63.	Find peak sales month and peak returned month for each Region.

```python
import pandas as pd

# 1. First check available columns
print("Available columns:", data.columns.tolist())

# Reset index to make OrderID a column for aggregation
data_temp = data.reset_index()

# Rename the column created by reset_index to 'OrderID' if it's named 'index'
# This handles cases where the original index didn't have a name.
if 'index' in data_temp.columns and 'OrderID' not in data_temp.columns:
    data_temp.rename(columns={'index': 'OrderID'}, inplace=True)

# 2. Identify which columns we have for returns/status
return_indicator = None
possible_status_columns = ['OrderStatus', 'Status', 'ReturnStatus', 'IsReturned', 'ReturnFlag']

for col in possible_status_columns:
    if col in data_temp.columns:
        return_indicator = col
        break

if return_indicator is None:
    # If no status column exists, we'll need to use another approach
    print("\nWarning: No order status column found. Using alternative methods...")

    # Alternative: Check if returns are indicated by negative amounts
    if 'TotalAmount' in data_temp.columns:
        print("Using negative amounts as return indicators")
        data_temp['IsReturn'] = data_temp['TotalAmount'] < 0
    else:
        raise KeyError("No order status column found and no alternative method available")
else:
    print(f"\nUsing column '{return_indicator}' to identify returns")
    data_temp['IsReturn'] = data_temp[return_indicator].astype(str).str.lower().str.strip().isin(['returned', 'refunded', 'return'])

# 3. Proceed with analysis (now using 'IsReturn' column)
data_temp['OrderDate'] = pd.to_datetime(data_temp['OrderDate'])
data_temp['YearMonth'] = data_temp['OrderDate'].dt.to_period('M')

# 4. Calculate metrics by region
monthly_region_stats = (
    data_temp.groupby(['Region', 'YearMonth'])
    .agg(
        TotalSales=('TotalAmount', 'sum'),
        TotalOrders=('OrderID', 'count'),
        ReturnedOrders=('IsReturn', 'sum')
    )
    .reset_index()
)

# 5. Find peak months
peak_sales_month = monthly_region_stats.loc[monthly_region_stats.groupby('Region')['TotalSales'].idxmax()]
peak_returned_month = monthly_region_stats.loc[monthly_region_stats.groupby('Region')['ReturnedOrders'].idxmax()]

print("\nPeak Sales Month per Region:")
print(peak_sales_month)

print("\nPeak Returned Month per Region:")
print(peak_returned_month)
```

64.	Identify cities that saw consistent monthly growth for 3+ months in a row.

```python
import pandas as pd

# 1. Prepare the data - ensure proper types
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['YearMonth'] = data['OrderDate'].dt.to_period('M')

# 2. Calculate monthly sales per city
monthly_sales = (
    data.groupby(['City', 'YearMonth'])['TotalAmount']
    .sum()
    .unstack(fill_value=0)
    .stack()
    .reset_index(name='MonthlySales')
)

# 3. Sort by City and YearMonth
monthly_sales = monthly_sales.sort_values(['City', 'YearMonth'])

# 4. Calculate month-over-month growth
monthly_sales['Growth'] = (
    monthly_sales.groupby('City')['MonthlySales']
    .pct_change()
    .mul(100)
    .round(2)
)

# 5. Identify growth streaks
def find_growth_streaks(group):
    streaks = []
    current_streak = 0
    for growth in group['Growth']:
        if growth > 0:  # Positive growth
            current_streak += 1
        else:
            if current_streak >= 3:
                streaks.append(current_streak)
            current_streak = 0
    # Check if streak continues to end
    if current_streak >= 3:
        streaks.append(current_streak)
    return pd.Series({
        'MaxStreak': max(streaks) if streaks else 0,
        'GrowthStreaks': streaks,
        'Has3MonthGrowth': any(streak >= 3 for streak in streaks) if streaks else False
    })

# 6. Apply to each city
city_growth = monthly_sales.groupby('City').apply(find_growth_streaks).reset_index()

# 7. Filter cities with 3+ month growth streaks
consistent_growth_cities = city_growth[city_growth['Has3MonthGrowth']]

# 8. Get detailed streak info
def get_streak_details(city):
    city_data = monthly_sales[monthly_sales['City'] == city].set_index('YearMonth').copy() # Set YearMonth as index and use copy
    city_data = city_data.sort_index() # Ensure index is monotonic

    streaks = []
    current_streak = 0
    start_month = None

    # Iterate through the index (YearMonth) and growth values
    for month, growth in city_data['Growth'].items():
        if growth > 0:
            current_streak += 1
            if current_streak == 1:
                start_month = month
        else:
            if current_streak >= 3:
                # Get the previous month Period
                end_month = month.asfreq('M') - 1 # Correct way to get previous period using asfreq and subtraction

                streaks.append({
                    'StartMonth': start_month,
                    'EndMonth': end_month,
                    'Months': current_streak,
                    'AvgGrowth': city_data.loc[start_month:end_month, 'Growth'].mean().round(2)
                })
            current_streak = 0
            start_month = None

    # Check if streak continues to end
    if current_streak >= 3:
        streaks.append({
            'StartMonth': start_month,
            'EndMonth': city_data.index[-1], # Last month in the data
            'Months': current_streak,
            'AvgGrowth': city_data.loc[start_month:, 'Growth'].mean().round(2)
        })

    return streaks

# 9. Create final result
results = []
for _, row in consistent_growth_cities.iterrows():
    city = row['City']
    streaks = get_streak_details(city)
    for streak in streaks:
        results.append({
            'City': city,
            'StartMonth': streak['StartMonth'],
            'EndMonth': streak['EndMonth'],
            'GrowthMonths': streak['Months'],
            'AverageGrowth%': streak['AvgGrowth'],
            'MaxStreak': row['MaxStreak']
        })

growth_report = pd.DataFrame(results).sort_values(['GrowthMonths', 'AverageGrowth%'], ascending=[False, False])

# 10. Display results
print("Cities with 3+ Months of Consistent Growth:")
print(growth_report.to_string(index=False))

# 11. Optional visualization
import matplotlib.pyplot as plt

top_cities = growth_report['City'].unique()[:5]  # Show top 5 for visualization

plt.figure(figsize=(12, 6))
for city in top_cities:
    city_data = monthly_sales[monthly_sales['City'] == city]
    plt.plot(city_data['YearMonth'].astype(str), city_data['MonthlySales'], label=city, marker='o')

plt.title('Monthly Sales Growth in Top Performing Cities')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

65.	Generate a YoY (Year-over-Year) growth percentage in TotalAmount for each Region

```python
import pandas as pd

# 1. Prepare the data
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['Year'] = data['OrderDate'].dt.year
data['Month'] = data['OrderDate'].dt.month

# 2. Calculate yearly sales by region
yearly_sales = (
    data.groupby(['Region', 'Year'])['TotalAmount']
    .sum()
    .unstack()
)

# 3. Calculate YoY growth
for year in range(yearly_sales.columns.min() + 1, yearly_sales.columns.max() + 1):
    prev_year = year - 1
    if prev_year in yearly_sales.columns:
        yearly_sales[f'YoY_Growth_{year}'] = (
            (yearly_sales[year] - yearly_sales[prev_year]) / yearly_sales[prev_year] * 100
        ).round(2)

# 4. Format results
result = yearly_sales.copy()
result = result.fillna('')  # Remove NaN for display

# 5. Display final YoY growth table
print("Year-over-Year Growth by Region (%):")
print(result.filter(like='YoY').to_string())

# 6. Visualization
import matplotlib.pyplot as plt

# Prepare data for plotting
plot_data = yearly_sales.filter(like='YoY').stack().reset_index()
plot_data.columns = ['Region', 'Year', 'YoY_Growth']
plot_data['Year'] = plot_data['Year'].str.replace('YoY_Growth_', '').astype(int)

plt.figure(figsize=(12, 6))
for region in plot_data['Region'].unique():
    region_data = plot_data[plot_data['Region'] == region]
    plt.plot(region_data['Year'], region_data['YoY_Growth'],
             label=region, marker='o', linestyle='-')

plt.title('Year-over-Year Growth by Region', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Percentage (%)', fontsize=12)
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
plt.grid(axis='y', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# 7. Detailed breakdown (optional)
print("\nDetailed Yearly Sales with YoY Growth:")
detailed_view = yearly_sales.copy()
detailed_view = detailed_view.applymap(lambda x: f"${x:,.2f}" if isinstance(x, (int, float)) else x)
print(detailed_view.to_string())
```

89.	After merge, create a master file with these columns: OrderID, OrderDate,
     CustomerName, Product, Category, Price, Quantity, TotalAmount, Returned, and the       AvgProductPrice and TotalCustomerOrders from lookup tables.


```python
import pandas as pd

# Load datasets
sales = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')
customers = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')
products = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')

# 1. First, identify common columns for merging
def find_common_col(df1, df2):
    """Find common columns between two DataFrames"""
    common = set(df1.columns) & set(df2.columns)
    return list(common) if common else None

# Find customer join key
cust_join_col = find_common_col(sales, customers)
if not cust_join_col:
    raise KeyError("No common columns found between sales and customers data")
cust_join_col = cust_join_col[0]  # Use first common column

# Find product join key
prod_join_col = find_common_col(sales, products)
if not prod_join_col:
    raise KeyError("No common columns found between sales and products data")
prod_join_col = prod_join_col[0]  # Use first common column

print(f"Merging customers using: {cust_join_col}")
print(f"Merging products using: {prod_join_col}")

# 2. Perform merges
try:
    # Merge sales with customers
    merged = pd.merge(
        sales,
        customers,
        on=cust_join_col,
        how='left'
    )

    # Merge with products
    master = pd.merge(
        merged,
        products,
        on=prod_join_col,
        how='left'
    )

    # 3. Create final columns (with fallbacks)
    final_columns = {
        'OrderID': 'OrderID' if 'OrderID' in master else None,
        'OrderDate': 'OrderDate' if 'OrderDate' in master else None,
        'CustomerName': 'CustomerName' if 'CustomerName' in master else cust_join_col,
        'Product': 'ProductName' if 'ProductName' in master else prod_join_col,
        'Category': 'Category' if 'Category' in master else None,
        'Price': 'Price' if 'Price' in master else 'UnitPrice',
        'Quantity': 'Quantity' if 'Quantity' in master else 'Qty',
        'TotalAmount': 'TotalAmount' if 'TotalAmount' in master else None,
        'Returned': 'Returned' if 'Returned' in master else 'IsReturned',
        'AvgProductPrice': 'AvgProductPrice' if 'AvgProductPrice' in master else None,
        'TotalCustomerOrders': 'TotalCustomerOrders' if 'TotalCustomerOrders' in master else None
    }

    # Filter out None values and keep only existing columns
    available_columns = {k:v for k,v in final_columns.items() if v in master.columns}
    master = master.rename(columns={v:k for k,v in available_columns.items()})[available_columns.keys()]

    # 4. Handle missing data
    for col in ['CustomerName', 'Product', 'Category']:
        if col in master.columns:
            master[col] = master[col].fillna(f'Unknown {col}')

    # Calculate missing numeric fields if needed
    if 'TotalAmount' not in master.columns and all(c in master.columns for c in ['Price', 'Quantity']):
        master['TotalAmount'] = master['Price'] * master['Quantity']

    if 'AvgProductPrice' not in master.columns and 'Price' in master.columns:
        master['AvgProductPrice'] = master.groupby('Product')['Price'].transform('mean')

    if 'TotalCustomerOrders' not in master.columns and 'OrderID' in master.columns:
        master['TotalCustomerOrders'] = master.groupby('CustomerName')['OrderID'].transform('nunique')

    # 5. Save and verify
    print("\nMaster file columns:", master.columns.tolist())
    print("Sample records:")
    print(master.head(3).to_string(index=False))

    master.to_csv('sales_master_file.csv', index=False)
    print("\nMaster file saved successfully!")

except Exception as e:
    print(f"Error during merge: {str(e)}")
    print("\nDebug Info:")
    print("Sales columns:", sales.columns.tolist())
    print("Customer columns:", customers.columns.tolist())
    print("Product columns:", products.columns.tolist())
```
